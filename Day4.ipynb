# Day 4
As far as the Elf has been able to figure out, you have to figure out which of the numbers you have appear in the list of winning numbers. The first match makes the card worth one point and each match after the first doubles the point value of that card.

For example:

Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53
Card 2: 13 32 20 16 61 | 61 30 68 82 17 32 24 19
Card 3:  1 21 53 59 44 | 69 82 63 72 16 21 14  1
Card 4: 41 92 73 84 69 | 59 84 76 51 58  5 54 83
Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11
In the above example, card 1 has five winning numbers (41, 48, 83, 86, and 17) and eight numbers you have (83, 86, 6, 31, 17, 9, 48, and 53). Of the numbers you have, four of them (48, 83, 17, and 86) are winning numbers! That means card 1 is worth 8 points (1 for the first match, then doubled three times for each of the three matches after the first).

Card 2 has two winning numbers (32 and 61), so it is worth 2 points.
Card 3 has two winning numbers (1 and 21), so it is worth 2 points.
Card 4 has one winning number (84), so it is worth 1 point.
Card 5 has no winning numbers, so it is worth no points.
Card 6 has no winning numbers, so it is worth no points.
So, in this example, the Elf's pile of scratchcards is worth 13 points.

```python
def split_string(lines):
    new_list=[]
    for line in lines:
        new_list.append(line.split('|'))
    return new_list
```

```python
def delete_card(elementos):
    new_list=[]
    aux=''
    for elemento in elementos:
        aux=elemento.split(':')
        new_list.append(aux[1])
    return new_list
```

```python
def to_int(listas):
    lista_numeros=[]
    numeros=[]
    for lista in listas:
        numeros=[]
#         if '' in lista:
#             print(lista)
#             lista.remove('')
            
#         if ' ' in lista:
#             print(lista)
#             lista.remove(' ')
        for numero in lista:
            num=int(numero)
            numeros.append(num)
        lista_numeros.append(numeros)
    return lista_numeros
```

```python
def remove_(line):
    new_line=[]
    for c in line:
        if c!='':
            new_line.append(c)
    return new_line
```

```python
def ret_lista_num(listas):
    new_list=[]
    sub_list=[]
    for lista in listas:
        sub_list=lista.split(' ')
        sub_list=remove_(sub_list)
        #print(sub_list)
        new_list.append(sub_list)
    return new_list
```

```python
def check_in_other(num1,num2):
    cont=0
    cards=0
    for num in num1:
        if num in num2:
            print(num)
            cards=cards+1
            cont=2**(cards-1)
    print('\nCont: ', cont,'\nNext')
    return cont   
```

```python
def compare_lists(num1,num2):
    n=len(num1)
    cont=0
    for i in range(0,n):
        cont+=check_in_other(num1[i],num2[i])
    return cont
```
