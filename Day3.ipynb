```python
def una_lista(numeros, posicion_inicial,posicion_final):
    lista_resultado=[]
    for numero, posicion_inicial, posicion_final in zip(numeros, posicion_inicial, posicion_final):

        # Creamos una tupla con el número, la posición inicial y la posición final
        tupla = [numero, posicion_inicial, posicion_final]

        # Agregamos la tupla a la lista resultado
        lista_resultado.append(tupla)

    return lista_resultado
```

```python
def una_lista2(numeros, posicion):
    lista_resultado=[]
    for numero, posicion in zip(numeros, posicion):

        # Creamos una tupla con el número, la posición inicial y la posición final
        tupla = [numero, posicion]

        # Agregamos la tupla a la lista resultado
        lista_resultado.append(tupla)

    return lista_resultado
```

```python
def nums_pos(mtx):
    pos_principio=[]
    numbers=[]
    pos_final=[]
    n_string=''
    for i,row in enumerate(mtx):
        for j,char in enumerate(row):
            if char.isdigit()==True:
                #print(char)
                n_string=n_string+char
                if (row[j-1].isdigit()==False and j>0) or j==0:
                    #print(char)
                    pos_principio.append([i,j])
                if (row[j+1].isdigit()==False and j<(len(row)-1))or j==(len(row)-1):
                    numbers.append(int(n_string))
                    n_string=''
                    pos_final.append([i,j])
                   
    return numbers,pos_principio,pos_final
```

```python
def ret_num(real):
    number_string=''
    num=[]
    pos=[]
    longitudes=[]
    for i,char in enumerate(real):
        #print(char)
        if char.isdigit():
            if i==0:
                pos.append(i)
            if real[i-1].isdigit()==False and i>0:
                pos.append(i)
            number_string=number_string+char
            number=int(number_string)
            long=len(number_string)
            if real[i+1].isdigit()==False and i<len(real):
                num.append(number)
                longitudes.append(long)
                number_string=''
    return num,longitudes,pos
```

```python
def digits_to_numbers(line):
    (numeros,long,posiciones)=ret_num(lines[0])
    #print('\nNumeros presentes: ', numeros, '\nPosiciones: ', posiciones, '\nLongitud c/n: ', long)
    lista_s=list(line)
    #print(lista_s)
    lista_m=lista_s
    for i,numero in enumerate(numeros):
        lista_m[posiciones[i]]=numero
        for char in range(long[i]):
            lista_m[posiciones[i]+char+1]='.'
    return lista_m
            
```

```python
def to_list(line):
    lista=list(line)
    return lista
```

```python
def to_matrix(lines):
    mtx=[]
    for line in lines:
        mtx.append(to_list(line))
    return(mtx)
```

```python
def check_symbol(char):
    yes=False
    symbols=['!', '"', '#', '$', '%', '&', "'", '(', ')', '*', '+', '-', '/', ':', ';', '<', '=', '>', '?', '@']
    for symbol in symbols:
        if char==symbol:
            yes=True
    return yes
```

```python
def list_nums(row,pos):
    n_s=[]
    for nums in range(pos,len(row)):
        if row[nums].isdigit()==True:
            n_s.append(row[nums])
    return n_s
```

```python
def list_to_number(lista):
    n_s=''
    for numbers in lista:
        n_s=n_s+numbers
    return int(n_s)
```
