Something is wrong with global snow production, and you've been selected to take a look. The Elves have even given you a map; on it, they've used stars to mark the top fifty locations that are likely to be having problems.

You've been doing this long enough to know that to restore snow operations, you need to check all fifty stars by December 25th.

Collect stars by solving puzzles. Two puzzles will be made available on each day in the Advent calendar; the second puzzle is unlocked when you complete the first. Each puzzle grants one star. Good luck!

You try to ask why they can't just use a weather machine ("not powerful enough") and where they're even sending you ("the sky") and why your map looks mostly blank ("you sure ask a lot of questions") and hang on did you just say the sky ("of course, where do you think snow comes from") when you realize that the Elves are already loading you into a trebuchet ("please hold still, we need to strap you in").

As they're making the final adjustments, they discover that their calibration document (your puzzle input) has been amended by a very young Elf who was apparently just excited to show off her art skills. Consequently, the Elves are having trouble reading the values on the document.

The newly-improved calibration document consists of lines of text; each line originally contained a specific calibration value that the Elves now need to recover. On each line, the calibration value can be found by combining the first digit and the last digit (in that order) to form a single two-digit number.

```python
def just_numbers(lines):
    new_line = [''] * len(lines)  # Crea una lista con len(lines) strings vacíos

    for (idx,line) in enumerate(lines):
        new_line[idx] = ""  # Crea una nueva cadena vacía
        for char in line:  # Recorre cada carácter de la cadena original
            if not char.isalpha():  # Si el carácter no es una letra, agrégalo a la nueva cadena
                new_line[idx] += char
    return new_line
```

```python
def two_digits(new_line):
    for (pos,line) in enumerate(new_line):
        if len(line)<2:
            new_line[pos]=line[0]+line[0]
        elif len(line)>=3:
            new_line[pos]=line[0]+line[len(line)-1]
    return new_line
```

```python
def to_ints(new_lines):
    ints = list(int(line) for line in new_lines)
    return ints
```

```python
def sumar(ints):
    total_sum = 0
    for int_ in ints:
        total_sum+=int_
    return total_sum
```

```python
def leters_to_numbers(lines):
    new_lines=[]
    for (idx, line) in enumerate(lines):
        ref = []
        if 'zero' in line:
            ref.append((line.find('zero'), 0))
        if 'one' in line:
            ref.append((line.find('one'), 1))
        if 'two' in line:
            ref.append((line.find('two'), 2))
        if 'three' in line:
            ref.append((line.find('three'), 3))
        if 'four' in line:
            ref.append((line.find('four'), 4))
        if 'five' in line:
            ref.append((line.find('five'), 5))
        if 'six' in line:
            ref.append((line.find('six'), 6))
        if 'seven' in line:
            ref.append((line.find('seven'), 7))
        if 'eight' in line:
            ref.append((line.find('eight'), 8))
        if 'nine' in line:
            ref.append((line.find('nine'), 9))

        new_lines.append(ref)
    return new_lines
    
```

```python
def numbers_to_numbers(lines):
    new_lines=[]
    for (idx, line) in enumerate(lines):
        ref = []
        if '0' in line:
            ref.append((line.find('0'), 0))
        if '1' in line:
            ref.append((line.find('1'), 1))
        if '2' in line:
            ref.append((line.find('2'), 2))
        if '3' in line:
            ref.append((line.find('3'), 3))
        if '4' in line:
            ref.append((line.find('4'), 4))
        if '5' in line:
            ref.append((line.find('5'), 5))
        if '6' in line:
            ref.append((line.find('6'), 6))
        if '7' in line:
            ref.append((line.find('7'), 7))
        if '8' in line:
            ref.append((line.find('8'), 8))
        if '9' in line:
            ref.append((line.find('9'), 9))

        new_lines.append(ref)
    return new_lines
```
